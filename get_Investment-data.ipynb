```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define the number of samples
num_samples = 10000

# Create the DataFrame with random data
data = {
    'CustomerID': range(1, num_samples + 1),
    'Age': np.random.randint(20, 70, num_samples),
    'Gender': np.random.choice(['Male', 'Female'], num_samples),
    'Marital_Status': np.random.choice(['Single', 'Married'], num_samples),
    'Household_Size': np.random.randint(1, 6, num_samples),
    'Education_Level': np.random.choice(['High School', 'Bachelor\'s', 'Master\'s', 'PhD'], num_samples),
    'Occupation': np.random.choice(['Engineer', 'Doctor', 'Sales', 'Professor', 'IT Manager', 'Analyst'], num_samples),
    'Income_Level': np.random.choice(['Low', 'Medium', 'High', 'Very High'], num_samples),
    'Total_Assets': np.random.randint(100000, 1000000, num_samples),
    'Liabilities': np.random.randint(50000, 300000, num_samples),
    'Net_Worth': np.random.randint(50000, 800000, num_samples),
    'Debt-to-Income': np.round(np.random.uniform(0.1, 0.5, num_samples), 2),
    'Preferred_Assets': np.random.choice(['Stocks', 'Bonds', 'Commodities', 'Real Estate', 'Mutual Funds', 'ETFs'], num_samples),
    'Sector_Preferences': np.random.choice(['Technology', 'Healthcare', 'Renewable Energy', 'Financial Services', 'Consumer Discretionary', 'Industrial'], num_samples),
    'Return_Expectation': np.random.randint(5, 15, num_samples),
    'Risk_Appetite': np.random.choice(['Conservative', 'Moderate', 'Aggressive'], num_samples),
    'Investment_Horizon': np.random.choice(['Short-term', 'Medium-term', 'Long-term'], num_samples),
    'GDP_Growth_Rate': np.round(np.random.uniform(2.0, 5.0, num_samples), 1),
    'Inflation_Rate': np.round(np.random.uniform(1.0, 3.0, num_samples), 1),
    'Interest_Rates': np.round(np.random.uniform(2.0, 5.0, num_samples), 1),
    'Stock_Indexes': np.random.choice(['S&P 500', 'NASDAQ', 'Russell 2000', 'Dow Jones', 'FTSE 100', 'Nikkei 225'], num_samples),
    'Compliance_With_Regulations': np.random.choice(['Yes', 'No'], num_samples),
    'Tax_Implications': np.random.choice(['Low', 'Medium', 'High'], num_samples),
    'KYC_Requirements': np.random.choice(['Yes', 'No'], num_samples),
    'ESG_Criteria': np.random.choice(['Yes', 'No'], num_samples),
    'Social_Responsibility': np.random.choice(['Low', 'Medium', 'High'], num_samples),
    'Corporate_Governance': np.random.choice(['Fair', 'Good', 'Excellent'], num_samples),
    'Asset_Allocation': np.random.randint(30, 80, num_samples),
    'Historical_Returns': np.random.randint(5, 15, num_samples),
    'Volatility': np.random.randint(10, 20, num_samples),
    'Correlation': np.round(np.random.uniform(0.1, 0.9, num_samples), 1),
    'Liquidity': np.random.choice(['High', 'Medium', 'Low'], num_samples),
    'Investment_Fees': np.round(np.random.uniform(1.0, 3.0, num_samples), 1),
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Save the DataFrame to an Excel file
df.to_excel('enhanced_investment_portfolio_data.xlsx', index=False)
```
